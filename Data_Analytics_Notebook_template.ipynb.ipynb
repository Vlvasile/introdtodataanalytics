```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('phone.csv')
```

```python
df.head()
```

```python
print("\nDataset shape:", df.shape)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
# no null great
print(df.describe())
```

```python
plt.figure(figsize=(15, 12))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Distribution of price ranges
plt.figure(figsize=(6, 4))
sns.countplot(x='price_range', data=df, palette="pastel")
plt.title("Distribution of Mobile Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.show()
```

```python
num_features = ['battery_power', 'ram', 'px_height', 'px_width', 'int_memory']

for feature in num_features:
    plt.figure(figsize=(7, 4))
    sns.boxplot(x='price_range', y=feature, data=df, palette="Set3")
    plt.title(f"{feature} vs Price Range")
    plt.xlabel("Price Range")
    plt.ylabel(feature)
    plt.show()

# Plot binary/categorical features against price range
bin_features = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']

for feature in bin_features:
    plt.figure(figsize=(6, 3))
    sns.countplot(x=feature, hue='price_range', data=df, palette="Set2")
    plt.title(f"{feature.capitalize()} Distribution by Price Range")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.legend(title="Price Range")
    plt.show()

# Pairplot for selected important features
selected = ['battery_power', 'ram', 'px_height', 'px_width', 'price_range']
sns.pairplot(df[selected], hue='price_range', palette="husl")
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```
